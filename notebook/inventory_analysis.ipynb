# 📊 Inventory Audit Analysis (SQL + Python)

## Análisis de inconsistencias en inventarios


## 🎯 Objetivo del proyecto

El objetivo de este proyecto es detectar inconsistencias entre el inventario registrado, las ventas y el inventario físico, con el fin de identificar posibles errores operativos y mejorar el control interno.

Este tipo de análisis es común en auditorías internas y permite prevenir pérdidas, errores de registro y problemas en la gestión de inventarios.

## 📂 Estructura de los datos

Se trabajó con tres tablas principales:

- **inventario_sistema**: contiene el stock registrado de cada producto
- **ventas**: registra las transacciones realizadas
- **inventario_fisico**: representa el conteo real del inventario

Estas tablas permiten comparar el stock esperado vs el real.

```python
# Importación de librerias utilizadas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Datos creados para el proyecto
data = {
    "producto": ["Laptop", "Mouse", "Teclado", "Monitor", "Impresora"],
    "stock_registrado": [50, 150, 100, 70, 30],
    "ventas_totales": [15, 30, 20, 15, 5],
    "stock_fisico": [30, 100, 70, 40, 20]
}

df = pd.DataFrame(data)

df["stock_esperado"] = df["stock_registrado"] - df["ventas_totales"]
df["diferencia"] = df["stock_esperado"] - df["stock_fisico"]

df
```

## 🔍 Análisis de inconsistencias

Se calcula la diferencia entre el stock esperado y el stock físico para identificar posibles discrepancias.

```python
df.sort_values(by="diferencia", ascending=False)
```

## 📊 Visualización de diferencias

Se grafican las diferencias para identificar rápidamente los productos con mayores inconsistencias.

```python
plt.figure()
plt.bar(df["producto"], df["diferencia"])
plt.title("Diferencias de Inventario por Producto")
plt.xlabel("Producto")
plt.ylabel("Diferencia")
plt.xticks(rotation=45)
plt.show()
```

## 📊 Interpretación

La gráfica muestra diferencias entre el inventario esperado y el inventario físico en varios productos.

Se observa que el producto **Mouse** presenta la mayor desviación, lo que indica posibles fallas en el control de inventarios o errores en el registro de ventas.

## 📌 Impacto en el negocio

Estas diferencias pueden generar:
- Pérdidas económicas
- Datos incorrectos para la toma de decisiones
- Problemas en auditorías internas

## 📈 Resultados

- 80% de los productos presentan inconsistencias
- Desviación máxima: 20 unidades (Mouse)

## 💡 Insights

- Se identificaron inconsistencias en 4 de 5 productos analizados.
- El producto "Mouse" presenta la mayor diferencia, lo que indica un posible problema en el registro de ventas o control de inventario.
- Estas discrepancias pueden representar pérdidas operativas o errores administrativos.

Este análisis permite enfocar auditorías en los productos con mayor desviación.

## ✅ Conclusión

El análisis permitió detectar diferencias relevantes entre el inventario esperado y el físico, evidenciando posibles fallas en los procesos de control interno.
